```python
def quitar_tildes(texto): # Codigo para quitar las tildes de un texto
    texto_normalizado = unicodedata.normalize('NFD', texto)
    texto_sin_tildes = ''.join(c for c in texto_normalizado if unicodedata.category(c) != 'Mn')
    return texto_sin_tildes
```

```python
texto_limpio0 = re.sub(r'[^A-Za-z0-9áéíóúÁÉÍÓÚñÑ\s]', '', luis_miguel) # Eliminamos carácteres especiales de un texto
texto_limpio1 = quitar_tildes(texto_limpio0)
texto_limpio1
```

```python
def eliminar_superindices(texto):
    return re.sub(r'(\b\w+\d+)', lambda m: m.group(0)[:-len(m.group(1))], texto)

texto_sin_numeros_finales = eliminar_superindices(texto_sin_preposiciones)
print(texto_sin_numeros_finales)
```
